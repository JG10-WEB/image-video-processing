PART A

```
#importing the required libraries
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```
# defining a function to count the number of times a pixel occurs 
#the function returns the pixel values and their count as lists
def histogram_image(img):
  [m,n]=img.shape
  count=[]
  pixels=[]
  for k in range(0,256):
    pixels.append(k)
    countt=0
    for i in range(m):
      for j in range(n):
        if img[i,j]==k:
          countt+=1
    count.append(countt)
  return (pixels,count)
```

```
#reading the image
img1=cv2.imread('/content/cameraman.tif',0)
type(img1)
```

```
#calling the function defined above
pixel1,count1=histogram_image(img1)
```

```
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.title('original image')
plt.imshow(img1,cmap='gray')
plt.subplot(1,2,2)
plt.stem(pixel1,count1)
```

PART C- Histogram Streching

Formula for Histogram Streching: 

S = T(r) = ((Smax - Smin)/(rmax - rmin))(r - rmin) + Smin

```
rmax=np.max(img3) # max pixel value in image
rmin=np.min(img3) # min pixel value in image
smax=255    # max pixel value in new image
smin=0      # min pixel value in image

ratio=(smax-smin)/(rmax-rmin)

m1,n1=img3.shape
```

```
#Creating an array of zeros that will be the new image 
img6=np.zeros((m1,n1),dtype=np.int)

for i in range(m1):
  for j in range(n1):
     img6[i,j]=ratio*(img3[i,j]-rmin)+smin 
cv2.imwrite('HISTOGRAMSTRECHEDIMAGE.png',img6)
```

```
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('original image')
plt.imshow(img3,cmap='gray',vmin=0,vmax=255)
plt.subplot(2,2,2)
plt.title('Histogram of low contrast image')
plt.stem(pixel3,count3)
plt.subplot(2,2,3)
plt.title('Image after histogram streching')
plt.imshow(img6,cmap='gray',vmin=0,vmax=255)
plt.subplot(2,2,4)
plt.title('Histogram of after streching image')
pixel6,count6 =histogram_image(img6)
plt.stem(pixel6,count6)
```

Inference: Here we have performed histogram streching on a low contrast image. We infer that the process of histogram streching has not altered the shape of the original image but has only increased the dynamic range of the image thus we obtain a histogram of an image that has pixels spread throughout that dynamic range. This image so obtained is a high contrast image and thus provides more information.

PART D - Histogram Equalization

```
# nk is the sum of all elements in list count4
nk=0
c=len(count3)
for i in range(0,c):
    nk = nk + count3[i]
nk
```

```
#Calculating the pdf of the taken list
pdf=[]
for i in range(0,c):
  pdf.append(count3[i]/nk)
```

```
#Calculating the cdf of the given list
cdf=[pdf[0]]
for i in range(1,c):
  update=cdf[i-1]+pdf[i]
  cdf.append(update)
```

```
#Multiplying the cdf by(L-1)
mul=[]
for i in range(0,c):
  mul.append(cdf[i]*(255))
```

```
#Rounding off the values in the mul list to whole values
import math
rounded=[]
for i in range(0,c):
  rounded.append(math.ceil(mul[i]))
```

```
#Counting the number of occurance of rouunded pixels
count0=[]
for k in rounded:
    countt1=0
    for i in range(c):
        if rounded[i]==k:
          countt1+=1
    count0.append(countt1)
```

```
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Histogram of high contrast image')
plt.stem(pixel3,count3)
plt.subplot(1,2,2)
plt.title('Histogram after equalization')
plt.stem(rounded,count0)
```

Inference:

After trying the 4 different contrasting images we have applied image equalization

Here we can see that the overall shape of the histogram has changed and has become equalized while in histogram steching the shape of the histogram remains the same only its dynamic range changes.
